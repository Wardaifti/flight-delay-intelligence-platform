```python
import dlt
from pyspark.sql.functions import *
from pyspark.sql.types import *
import math

base_path = "/Volumes/workspace/default/my_volume/flight_project"

# Bronze paths
bronze_flights_path = f"{base_path}/bronze/flights"
bronze_weather_path = f"{base_path}/bronze/weather"

# ========================
# BRONZE LAYER
# ========================

@dlt.table(
    name="bronze_flights",
    comment="Raw flight data from OpenSky Network"
)
@dlt.expect("valid_icao", "icao24 IS NOT NULL")
@dlt.expect("valid_position", "latitude IS NOT NULL AND longitude IS NOT NULL")
@dlt.expect_or_drop("valid_altitude", "baro_altitude IS NOT NULL")
def bronze_flights():
    return spark.read.format("delta").load(bronze_flights_path)


@dlt.table(
    name="bronze_weather",
    comment="Raw weather data from OpenWeather API"
)
@dlt.expect("valid_airport", "airport_code IS NOT NULL")
@dlt.expect("valid_temp", "temperature IS NOT NULL")
def bronze_weather():
    return spark.read.format("delta").load(bronze_weather_path)
```
